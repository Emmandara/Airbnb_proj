```python
#replace these placeholders with np.nan using the .replace() method
import numpy as np
listings.replace(['?', 'NULL', 'NA', ''], np.nan, inplace=True)
```

```python
#List all duplicates in name column
duplicates =listings[listings.duplicated(['name'])]
print(duplicates)
```

```python
#save name duplicates to csv for further inspection
duplicates.to_csv("duplicates.csv", index=False)
```

```python
#remove name duplicates
listings_no_duplicates = listings[~listings.duplicated(subset="name", keep=False)]
listings.sample(10)
```

```python
#save new csv with no duplicates
listings_no_duplicates.to_csv("cleaned_no_name_duplicates.csv", index=False)
```

```python
import pandas as pd
cleaned = pd.read_csv("cleaned_listings_df.csv", low_memory=False)
cleaned.shape
```

```python
cleaned.head(5)
```

```python
#filter listings with a rating of 5.0
cleaned_filtered = cleaned[cleaned["review_scores_rating"] >= 5]
cleaned_filtered.head(5)
```

```python
#order by desc to show 5.0 ratings with the most reviews
cleaned_result = cleaned_filtered.sort_values(by="number_of_reviews", ascending=False)
cleaned_result.head(5)
```

```python
#save top 5 as a csv file for reference
cleaned_result.to_csv("top5_listings_df.csv", index=False)
```

```python
cleaned_listings = cleaned.copy()

# Clean price: remove $ and commas and convert to float
cleaned['price_clean'] = (cleaned['price'].replace(r'[$,]', '', regex=True).astype(float))
# Simplify bathrooms_text -> numeric bathrooms
cleaned['bathrooms_num'] = (cleaned['bathrooms_text'].str.extract(r'(\d+\.?\d*)')[0].astype(float))
```

```python
cleaned_listings.head(5)
```

```python
cols_to_keep = [
        'id','name','host_id','host_name','host_is_superhost',
        'neighbourhood_cleansed','neighbourhood_group_cleansed','neighborhood_overview',
        'latitude','longitude','property_type','room_type','accommodates','bedrooms','beds','bathrooms_num',
        'price_clean','number_of_reviews','review_scores_rating','reviews_per_month'
]
clean_listings = cleaned_listings[cols_to_keep]
print(f"After selecting columns: {clean_listings.shape}")
```
